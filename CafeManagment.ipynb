```python
import pandas as pd
from datetime import datetime
import os
from time import sleep

# Order Management System
class BakeryOrderManagement:
    def __init__(self):
        self.orders = pd.DataFrame(columns=["Customer Name", "Item", "Quantity", "Order Date" , "Order Status"])

    def add_order(self):
        customer_name = input("Enter customer name: ")
        item = input("Enter item ordered: ")
        quantity = int(input("Enter quantity: "))
        order_date = datetime.now().strftime("%Y-%m-%d")
        new_order = pd.DataFrame([[customer_name, item, quantity, order_date, 0]],
                                 columns=["Customer Name", "Item", "Quantity", "Order Date" ,"Order Status"])
        self.orders = pd.concat([self.orders, new_order], ignore_index=True)
        print("Order added successfully!")
        sleep(2)

    def order_completed(self ):
      order_id = int(input(("Enter order ID of the given order: ")))-1
      if order_id >= len(self.orders):
        print("Order does not exist")
      else:
        self.orders.at[order_id , "Order Status"] = 1
        print(f"Order no. {order_id+1} has been served.")

      sleep(3)

    def view_orders(self):
        if self.orders.empty:
            print("No orders available.")
        else:
            print(self.orders)

        sleep(5)

    def update_order(self):
        order_id = int(input("Enter order ID to update: "))-1
        if order_id >= len(self.orders):
            print("Order not found.")
        else:
            print("Updating Order ID:", order_id)
            self.orders.at[order_id, "Item"] = input("Enter new item: ")
            self.orders.at[order_id, "Quantity"] = int(input("Enter new quantity: "))
            print("Order updated successfully!")
            sleep(3)

    def save_to_excel(self):
        self.orders.to_csv("bakery_orders.csv", index=False)
        print("Orders saved to 'bakery_orders.csv'.")
        sleep(3)
```

```python

order_system = BakeryOrderManagement()

while True:
    # sleep(1)
    os.system('cls')
    print("\n1. Add Order\n2. View Orders\n3. Update Order\n4. Order completed\n5. Save to Excel\n6. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
        order_system.add_order()
    elif choice == '2':
        order_system.view_orders()
    elif choice == '3':
        order_system.update_order()
    elif choice == '4':
        order_system.order_completed()
    elif choice == '5':
        order_system.save_to_excel()
    elif choice == '6':
        break
    else:
        print("Invalid choice. Please try again.")
```
